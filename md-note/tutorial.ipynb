```python
athlete = pd.DataFrame([["Max", "Jakarta", 3.3], ["James", "Surabaya", 3.7], ["Bryan", "semarang", 3.5], ["Rafael", "Bekasi", 4.0]], columns=["name", "born", "gpa"])
athlete
```

```python
athlete[athlete['name'].str.contains("max|bryan", case=False, regex=True)] # regex = symbols like "|" which means OR
```

```python
athlete[athlete["born"].isin(["Jakarta", "Bekasi"]) & (athlete["name"].str.contains("max", case=False))]
```

```python
# athlete[#COLUMN].str.startswith("max") is also a function
athlete.query("name == 'Max' and born == 'Jakarta'")
```

```python
athlete['old_univ'] = "CUHK"
athlete.loc[[1,3],"univ"] = "HKUST"
athlete
```

```python
#another method, using numpy
import numpy as np
athlete['new_univ'] = np.where(athlete['gpa']>3.5, "MIT", "NUS")
athlete
```

```python
athlete.drop(columns=['univ'])
```

```python
athlete['scholarship'] = np.where(athlete["gpa"]<=3.5, "no", "yes")
athlete.drop(columns=["old_univ"])
athlete
```

```python
athlete = athlete.drop(columns=["old_univ"])
athlete = athlete.drop(columns=["univ"])
athlete
```

```python
athlete.rename(columns={"new_univ":"univ"}, inplace=True)
```

```python
athlete["born_date"] = ['2007-12-12', '2007-11-11', '2007-10-10', '2007-09-09']
athlete
```

```python
athlete["year_born"] = athlete["born_date"].str.split("-").str[0]
athlete
```

```python
athlete["born_datetime"] = pd.to_datetime(athlete["born_date"], format="%Y-%m-%d")
athlete
```

```python
# multiple conditions using apply lambda or  a def
athlete['level'] = athlete['gpa'].apply(lambda x: "High" if x > 3.6 else ("Average" if 3.4 < x <= 3.6 else "low"))
athlete
```

```python
new_info = pd.DataFrame([[175, 70], [178, 72], [170, 68], [168, 60]], columns = ["height", "weight"], index=["Max", "James", "Bryan", "Rafael"])
new_info
```

```python
#matching left_on and right_on
athlete_new = pd.merge(athlete, new_info, left_on="name", right_on=new_info.index, how="left")
athlete_new
```

```python
add_on = pd.DataFrame([["Blake", "USA", 2.5], ["Tyrone", "Jamaica", 3.0], ["Ben", "England", 2.9]], columns=["name", "born", "gpa"])
```

```python
students = pd.concat([athlete_new, add_on]) 
students
```

```python
students.fillna(students['year_born'].interpolate)
students
```

```python
students.groupby(['univ','level']).agg({'gpa':'mean','height':'mean'})
```

```python
sports = pd.DataFrame([['volleyball', 'Max'], ['basketball', 'James'], ['basketball', 'Bryan'], ['volleyball', 'Rafael'], ['volleyball', 'Blake'], ['basketball', 'Tyrone'], ['volleyball', 'Ben']], columns=["sports", 'name'])
```

```python
students = pd.merge(students, sports, left_on='name', right_on='name', how="left")
```

```python
pivot = students.pivot(columns='sports', index='name', values='gpa')
pivot
```

```python
pivot.mean(axis=1)
```

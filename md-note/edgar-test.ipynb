```python
import edgar
import pandas as pd
edgar.set_identity("sachiomaximilliano166@gmail.com")
nvda = edgar.Company("SPGI")
tenk = nvda.get_filings(form="8-K")
available_filings = []
for f in tenk:
    if '2.02' in f.items:
        text = f.obj().text()
        available_filings.append(text)

```

```python

lower_text = available_filings[-1].lower()
idx = lower_text.find('guidance')
start = max(0, idx - 200)
end = min(len(lower_text), idx + 500)
snippet = lower_text[start:end]
print(snippet)
```
